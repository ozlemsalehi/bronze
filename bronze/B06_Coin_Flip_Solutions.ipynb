<h3> Task 1: Simulating FairCoin in Python</h3>

Flip a fair coin 100 times. Calcuate the total number of heads and tails, and then compare them.

Do the same experiment 1000 times.

Do the same experiment 10,000 times.

Do the same experiment 100,000 times.

Do your results get close to the ideal case (the numbers of heads and tails are the same)?

<h3>Solution</h3>

```python
from random import randrange

for experiment in [100,1000,10000,100000]:
    heads = tails = 0
    for i in range(experiment):
        if randrange(2) == 0: heads = heads + 1
        else: tails = tails + 1
    print("experiment:",experiment)
    print("the ratio of #heads/#tails is",(heads/tails),"heads =",heads,"tails = ",tails)
    print() # empty line
```

<h3> Task 2: Simulating BiasedCoin in Python</h3>

Flip the following biased coin 100 times. Calcuate the total numbers of heads and tails, and then compare them.

Do the same experiment 1000 times.

Do the same experiment 10,000 times.

Do the same experiment 100,000 times.

Do your results get close to the ideal case $ \mypar{ \dfrac{ \mbox{# of heads} }{ \mbox{# of tails} } = \dfrac{0.6}{0.4} = 1.50000000 } $?

<h3>Solution</h3>

```python
from random import randrange

# let's pick a random number between {0,1,...,99}
# it is expected to be less than 60 with probability 0.6
#     and greater than or equal to 60 with probability 0.4

for experiment in [100,1000,10000,100000]:
    heads = tails = 0
    for i in range(experiment):
        if randrange(100) <60: heads = heads + 1 # with probability 0.6
        else: tails = tails + 1 # with probability 0.4
    print("experiment:",experiment)        
    print("the ratio of #heads/#tails is",(heads/tails),"heads =",heads,"tails = ",tails)
    print() # empty line
```

<h3> Task 3</h3>

Write a function to implement the described biased coin,

The inputs are integers $ N >0 $ and $ 0 \leq B < N $.

The output is either "Head" or "Tail".

<h3>Solution</h3>

```python
def biased_coin(N,B):
    from random import randrange
    random_number = randrange(N)
    if random_number < B:
        return "Head"
    else:
        return "Tail"
```

<h3> Task 4</h3>

We use the biased coin defined in Task 3. 

(You may use the one given in the solution.)

We pick $ N $ as 101.

We determine $ B $ randomly, and we do not check its value immedeately.

The task is to guess the bias by using the biased coin at most 500 times.

Check the real bias, and calculate the error.

<h3>Solution</h3>

```python
def biased_coin(N,B):
    from random import randrange
    random_number = randrange(N)
    if random_number < B:
        return "Head"
    else:
        return "Tail"
```

```python
from random import randrange
N = 101
B = randrange(100)
```

```python
total_tosses = 500
the_number_of_heads = 0
for i in range(total_tosses):
    if biased_coin(N,B) == "Head":
        the_number_of_heads = the_number_of_heads + 1

my_guess =  the_number_of_heads/total_tosses        
real_bias = B/N
error = abs(my_guess-real_bias)/real_bias*100 

print("my guess is",my_guess)
print("real bias is",real_bias)
print("error (%) is",error)
```
