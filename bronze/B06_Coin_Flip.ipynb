<h2> Coin Flipping </h2>


A coin has two sides: <i>Head</i> and <i>Tail</i>.



If our coin is fair, then the probabilities of obtaining a Head or a Tail are equal:

$ Pr(Head) = Pr(Tail)= \dfrac{1}{2} = 0.5 $.
